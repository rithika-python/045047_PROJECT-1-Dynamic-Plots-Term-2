```python
import pandas as pd
import plotly.express as px

# Assuming your CSV file has columns: 'year', 'population', 'states'
df = pd.read_csv("us-population-2010-2019-reshaped.csv")

fig = px.bar(
    df,
    x="states_code",
    y="population",
    animation_frame="year",  # Specify the 'year' column for animation
    color="states",
    hover_name="states",
    labels={"population": "Population", "year": "Year", "states_code": "States Code"},
    title="US Population Over Time",
)

fig.update_layout(
    xaxis_title="States",
    yaxis_title="Population",
    showlegend=True,
    legend_title_text="States",
    yaxis=dict(
        range=[0, 40000000],  # Adjust the range of the y-axis based on your preference
    ),
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Assuming your CSV file has columns: 'year', 'population', 'states'
df = pd.read_csv("us-population-2010-2019-reshaped.csv")

# Choose the 5 states you are interested in
selected_states = ["Kansas", "Iowa", "New York", "Idaho", "Mississippi"]

# Filter the DataFrame for the selected states
filtered_df = df[df["states"].isin(selected_states)]

# Set a base year for comparison
base_year = 2010

fig = go.Figure()

for state in selected_states:
    state_data = filtered_df[filtered_df["states"] == state]
    population_diff = state_data["population"] - state_data[state_data["year"] == base_year]["population"].values[0]
    trace = go.Scatter(
        x=state_data["year"],
        y=population_diff,
        mode='lines+markers',
        name=state
    )
    fig.add_trace(trace)

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Population Difference (vs. {})".format(base_year),
    title="Change in Population Over Time (Selected States)",
    showlegend=True
)

fig.show()
```
